```python
import pandas as pd
```

```python
# Open file    
fileHandler = open ("vale.txt", "r")
# Get list of all lines in file
listOfLines = fileHandler.readlines()
# Close file 
fileHandler.close()
```

```python
title = []
date = []
for i in range(len(listOfLines)):
    if i%2 == 0:
        title.append(listOfLines[i])
    else:
        date.append(listOfLines[i])
```

```python
remove = '0123456789%()-'
for i in range(len(title)):
    for j in range(len(remove)):
        title[i] = title[i].replace(remove[j],"")
```

```python
for i in range(len(title)):
        title[i] = title[i].replace("R$","")
```

```python
for i in range(len(title)):
        title[i] = ' '.join(title[i])
```

```python
for i in range(len(date)):
        date[i] = date[i][0]
```

```python
for i in range(len(date)):
        title[i] = title[i].split()
```

```python
lista = ['Exibição']
#,'Exibição'
#'Seg,','Ter,','Qua,','Qui,','Sex,','Sáb,','Dom,'
for i in range(len(date)):
    if date[i][0] in lista:
        date[i] = '-'.join([date[i][2],date[i][3],date[i][4]])
        
```

```python
title
```

```python
df= pd.DataFrame()
```

```python
df['Titulo'] = title
```

```python
df['Data'] = date
```

```python
df.to_csv('Vale',index = False)
```
